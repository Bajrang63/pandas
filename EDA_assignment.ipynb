  How does the death rate vary with age

```python
# Group data by age and calculate death rate
grouped_data = df.groupby('age')['DEATH_EVENT'].mean()

# Create a bar chart
plt.bar(grouped_data.index, grouped_data.values)
plt.xlabel('Age')
plt.ylabel('Death Rate')
plt.title('Death Rate by Age')
plt.show()
```
